```python
import os
import numpy as np
import pandas as pd
from sklearn.metrics import f1_score, roc_auc_score, matthews_corrcoef
```

```python
df = pd.read_excel('./CPI check_public_test_label.xlsx')
df = df.drop(columns=['index'])
df
```

```python
df['S01'] = (df['(S01)'] > 0.5).astype(int)
df['S02'] = (df['Baseline00(S02)'] > 0.5).astype(int)
df['S03'] = (df['Baseline00_vali_Loss(S03)'] > 0.5).astype(int)
df['S04'] = (df['Baseline00_CustomLabel(S04)'] > 0.5).astype(int)
df['BL01'] = (df['Baseline01'] > 0.5).astype(int)
df['BC01'] = (df['Baseline01_CustomLabel'] > 0.5).astype(int)
df
```

```python
df.columns
```

## CPI Score

---

### F1 average = `binary`(default) / AUROC average = `macro`(default)

```python
predict_col = [('(S01)', 'S01'), ('Baseline00(S02)', 'S02'), ('Baseline00_vali_Loss(S03)', 'S03'), ('Baseline00_CustomLabel(S04)', 'S04'), ('Baseline01', 'BL01'), ('Baseline01_CustomLabel', 'BC01')]

print("F1 score average = binary(default)")
print("AUROC score average = macro(default)\n")

for prob, pred in predict_col:
    # F1 Score ê³„ì‚°
    f1 = f1_score(df['label'], df[pred])

    # AUROC ê³„ì‚°
    auroc = roc_auc_score(df['label'], df[prob])

    # MCC ê³„ì‚°
    mcc = matthews_corrcoef(df['label'], df[pred])

    # CPI ê³„ì‚°
    cpi = 0.25 * f1 + 0.25 * auroc + 0.5 * mcc

    print(f"====== CPI estimate ===============\n{prob} | {pred}")
    print(f"- F1 Score: {f1}")
    print(f"- AUROC: {auroc}")
    print(f"- MCC: {mcc}")
    print(f">>> CPI: {cpi}")
```
